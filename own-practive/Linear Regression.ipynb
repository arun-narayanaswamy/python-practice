# Plot individuals

```python
import matplotlib.pyplot as plt
```

```python
def plot_residuals(model_untrained, X, y,random_state=None):
    X_train,y_train,x_test,y_test=train_test_split(X,y,
                                                   random_state=random_state)
    model=model_untrained.fit(X_train,y_train)
    fitted=model.predict(x_test)
    residual=fitted-y_test
    return plt.scatter(fitted,residual)
```

```python
def plot_residuals(model_untrained, X, y, random_state=None):

    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        random_state=random_state)
    model = model_untrained.fit(X_train, y_train)

    fitted = model.predict(X_test)
    residuals = fitted - y_test

    return plt.scatter(fitted, residuals)
```

```python
plot_residuals(LinearRegression(), X, Y)
```

```python
def score_model(estimator,X,Y):
    scores=[]
    test_scores=[]
    
    for i in range(1000):
        X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
        lr=estimator().fit(X_train,Y_train)
        scores.append(lr.score(X_train,Y_train))
        test_scores.append(lr.score(X_test,Y_test))
    scores=pd.DataFrame({'training':scores,'testing':test_scores})   
    return scores
```

```python
#score model on initial value of X and Y
score_model(LinearRegression,X,Y).describe()
```
