```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
iris_data=load_iris()
type(iris_data.data)
```

```python
print(iris_data.feature_names)
```

```python
df=pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
df.head(7)
```

```python
df.columns=[name[:-5].replace(' ','_') for name in df.columns]
```

```python
df.head()
```

```python
df['label']=iris_data.target_names[iris_data.target]
```

```python
df.head()
```

```python
df.describe()
```

```python
#use seaborn library for pairplot
sns.pairplot(data=df,hue='label')
```

```python
X=df.drop('label',axis=1).values
Y=df['label'].values
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=26)
```

```python
lgr=LogisticRegression().fit(X_train,Y_train)
```

```python
lgr.score(X_test,Y_test)
```

```python
lgr.score(X_train,Y_train)
```

```python
lnr=LinearRegression().fit(X_train,Y_train)
```

```python
lgr.predict(X_test)
```
